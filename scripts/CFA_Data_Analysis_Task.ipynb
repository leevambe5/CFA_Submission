4. Using a plotting library / tool display the daily number of tweets from ;

#### a. The top 2 category types within the datasets ranked by volume of tweets. 

```python
# Value counts method to view volume of tweets by category type
df['account_category'].value_counts()
```

```python
# Pivot table to view all tweets by category type and by date
df_pivot = df.pivot_table(index='date', columns='account_category', values=['tweet_id'], aggfunc='count')
df_pivot.head()
```

```python
# Replacing null values with zeros
df_pivot = df_pivot.replace(np.nan,0)
df_pivot
```

```python
# Displaying pivot table columns
df_pivot.columns
```

```python
df_top_cats = df_pivot.select(lambda x: x[1] in ['LeftTroll', 'RightTroll'], axis=1)
```

```python
df_top_cats.head(2)
```

```python
# Line plot to show volume of tweeets for the top 2 categories by date
df_top_cats.plot(kind='line',figsize=(18,8),title="Top Categories Tweets Trends");
```

#### b. The 3rd and 4th category types ranked by volume of tweets

```python
df_3rd_4th_cats = df_pivot.select(lambda x: x[1] in ['HashtagGamer', 'NewsFeed'], axis=1)
```

```python
# Line plot to show volume of tweeets for the 3rd and 4th categories by date
df_3rd_4th_cats.plot(kind='line',figsize=(18,8),title="3rd & 4th Categories Tweets Trends");
```

#### What are some of the characteristics you would consider when identifying fake or suspicious accounts on twitter?

* Excessive Duplicate Tweets
* Has Not Tweeted in Years
* Less personal information on the profile/ degree of anonymity 
* Activity per day
* Massive amplification
* Handle contains alphanumeric scrambles

* #### Using the data provided, identify one account that you suspect might be automated and flag some of the bot-like traits you identified.

```python
# Value count of duplicate tweets
duplicate_tweets = df[df.duplicated(['content','author'])].sort_values("content")
```

```python
duplicate_tweets['author'].value_counts()
```

```python
# Viewing all tweets done by AMELIEBALDWIN
AMELIEBALDWIN_tweets = duplicate_tweets[duplicate_tweets.loc[:,'author']=='AMELIEBALDWIN'].sort_values("date")
AMELIEBALDWIN_tweets
```

```python
# Counting the number of occurences of each tweet posted by AMELIEBALDWIN
AMELIEBALDWIN_tweets['content'].value_counts().to_frame()
```

```python
# Categorizing AMELIEBALDWIN tweets count by date
AMELIEBALDWIN_tweets['date'].value_counts()
```

```python
# Viewing AMELIEBALDWIN tweets done on 2017-02-08
AMELIEBALDWIN_tweets[AMELIEBALDWIN_tweets.loc[:,'date'].astype(str)=='2017-02-08'].sort_values("date_col")
```

#### Import Libraries

```python
import tweepy
import numpy as np 
import pandas as pd
import getpass
```

#### Parsing twitter developer account credentials

```python
consumer_key = getpass.getpass("Enter Consumer Key : ") 
consumer_key_secret = getpass.getpass("Enter Consumer Secret : ")
access_token = getpass.getpass("Enter Access Token : ") 
access_token_secret = getpass.getpass("Enter Access Token Secret : ") 
```

```python
# creating authentication for accessing Twitter
auth = tweepy.OAuthHandler(consumer_key, consumer_key_secret)
auth.set_access_token(access_token, access_token_secret)

# initializing Tweepy API
api = tweepy.API(auth,wait_on_rate_limit=True)
```

#### Data Collection

```python
# initialzing hashtag phrase and list to store our search results
keyword = "#HumanTrafficking"
tweet_list=[]
for tweet in tweepy.Cursor(api.search,q=keyword,count=200,tweet_mode='extended').items(10000):
    # labelling retweets as 1 and non retweets as 0 based on the first characters of the tweet
    if tweet.full_text.startswith("RT @") == True:
        rt_test = 1
    else:
        rt_test = 0
# appending search results into a list
    tweet_data = [[tweet.user.id,tweet.created_at,tweet.user.name,tweet.user.screen_name,tweet.user.followers_count,tweet.lang,tweet.user.location,rt_test,tweet.full_text]]
    for i in tweet_data:
        tweet_list.append(i)
tweet_list
```

#### DataFrame building for the search results

```python
# dataframe to hold input from the tweet_list
tweets_df = pd.DataFrame(tweet_list, 
                    columns=['user_id','create_date','name','screen_name','followers_count','language','location','retweet','content'])
tweets_df
```
